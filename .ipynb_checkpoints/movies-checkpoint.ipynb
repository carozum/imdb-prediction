### Revue des doublons

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'movies_df' is your DataFrame and it's already loaded
# Automatically select columns that have dtype 'object' or 'category'
categorical_columns = movies_df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)


```

```python

for col in categorical_columns:
    # It's a good practice to limit the number of categories to plot
    # This check skips columns with too many categories for readability
    if movies_df[col].nunique() <= 40:  # Adjust this threshold as necessary
        # Set up the matplotlib figure
        plt.figure(figsize=(10, 6))
        
        # Create the boxplot
        p = sns.boxplot(x="imdb_score", y=col, data=movies_df, saturation=.6, fliersize=10.,
                        palette=["lightgray", sns.color_palette()[1], "grey", "darkblue"])
        
        # Set the labels and title
        p.set(xlabel="IMDb Score", ylabel=col.capitalize(), xticks=range(0, 11))
        p.set_title(f"IMDb Score Distribution by {col.capitalize()}")
        
        # Show plot
        plt.show()
```

```python
p = sns.boxplot("imdb_score", "color", data = movies_df, saturation = .6, fliersize = 10., palette = ["lightgray", sns.color_palette()[1], "grey", "darkblue"])
_ = p.set(xlabel = "imdb_score", ylabel = "country of movie", xticks = range(11))
```

## CLEANING

### Suppression des lignes sans IMDB Score (--> notre variable cible)

```python
data_cleaned = movies_df.dropna(subset=['imdb_score'])
```

### Remplissage des manquants avec la mediane des valeurs

```python
data_cleaned['duration'].fillna(data_cleaned['duration'].median(), inplace=True)
```

```python
# Suppression de colonnes qui ne sont pas pertinentes pour l'analyse ou qui ont trop de valeurs manquantes
# (Ceci est juste un exemple, à ajuster selon vos besoins)
columns_to_drop = ['aspect_ratio', 'content_rating']
data_cleaned.drop(columns=columns_to_drop, inplace=True)

# Vérification après nettoyage
print(data_cleaned.info())
```

## ETUDE DES CORRELATIONS ET CHOIX DES COLONNES

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer la matrice de corrélation entre les variables numériques
corr = data_cleaned.corr()

# Configurer la taille de la figure matplotlib
plt.figure(figsize=(12, 10))

# Créer un heatmap pour visualiser la matrice de corrélation
# 'annot = True' affiche les valeurs des coefficients de corrélation
# 'cmap' définit la palette de couleurs. Par exemple, 'coolwarm', 'viridis', 'plasma', ...
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')

# Ajouter des titres pour une meilleure lisibilité
plt.title('Matrice de Corrélation')
plt.xticks(rotation=45, ha="right")  # Rotation des labels de l'axe x pour une meilleure lisibilité
plt.yticks(rotation=0)  # Garder les labels de l'axe y horizontaux

plt.show()
```
